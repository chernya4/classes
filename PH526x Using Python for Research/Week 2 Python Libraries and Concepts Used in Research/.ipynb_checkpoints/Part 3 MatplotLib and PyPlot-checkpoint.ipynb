# Introduction to Matplotlib and Pyplot #

PyPlot is a collection of functions that make matplotlib work like Matlab.

Pyplot provides a **state machine interface** to the matplotlib library. - Think of this as a process hwere you create figures one at a time and all commands affect the current figure and the current plot. 

**Keyword Argument** - argument which is supplied to the function by explicitly naming each parameter and specifying its value.

```python
import matplotlib.pyplot as plt
plt.plot([0,1,4,8,16]) # Python returns a matplotlib object.
# this ojbect is just those lines being printed out. 
```

```python
# Semicolon will force object not to be printed
plt.plot([0,1,4,8,16]); # Python returns a matplotlib object.
```

```python
import numpy as np
# Createa a vector x
x = np.linspace(0, 10, 20)
y = x ** 2 # Where all x is raised to second power.
```

```python
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2 # Where all x is raised to second power.
y2 = x ** 1.5
plt.plot(x, y1, "bo-") #blue circles and solid line
```

```python
plt.plot(x, y1, "bo-", linewidth = 2, markersize = 12) #blue circles and solid line
```

```python
plt.plot(x, y2, "gs-")
```

```python
plt.plot([0,1,2],[0,1,4],"rd-")
```

# Customizing Your Plots #
- legend() - add legend
- axis() - adjust axes
- xlabel() - set axis label
- ylabel() - set axis label
- savefig() - save figure
- **plt knows LaTeX**!!!!

```python
plt.plot(x, y1, "bo-", linewidth = 2, markersize = 12, label = "First") #blue circles and solid line
plt.plot(x, y2, "gs-", linewidth = 2, markersize = 12, label = "Second") #blue circles and solid line
plt.xlabel("$X$")
plt.ylabel("$Y$")
xmin = -0.5
xmax = 10.5
ymin = -5
ymax = 105
plt.axis([xmin, xmax, ymin, ymax])
plt.legend(loc = "upper left")
plt.savefig("myplot.pdf")
```

# Plotting Using Logarithmic Axes #

- semilogx() plots the x-axes on a log scale and the y in the original scale.
- semilogy() plots the y-axes on the log scale and hte x in the original scale.
- loglog() plots both x and y on the logarithmic scales.

y = x^a

Log transform of these axes:

logy = a * logx

**To Make log log axes evenly spaced** you can gereate numbers that are evenly spaced on a logarithmic scale.Using the **np.logspace** function.

```python
plt.loglog(x, y1, "bo-", linewidth = 2, markersize = 12, label = "First") #blue circles and solid line
plt.loglog(x, y2, "gs-", linewidth = 2, markersize = 12, label = "Second") #blue circles and solid line
plt.xlabel("$X$")
plt.ylabel("$Y$")
xmin = -0.5
xmax = 10.5
ymin = -5
ymax = 105
plt.axis([xmin, xmax, ymin, ymax])
plt.legend(loc = "upper left")
plt.savefig("myplot.pdf")
```

```python
# MAKING IT EVENLY SPACED IWHT THE np.logspace() ##

x = np.logspace(-1,1, 40) ## first point is going to be 0.1
#endpoint will be 10
y1 = x **2
y2 = x**1.5
plt.loglog(x, y1, "bo-", linewidth = 2, markersize = 12, label = "First") #blue circles and solid line
plt.loglog(x, y2, "gs-", linewidth = 2, markersize = 12, label = "Second") #blue circles and solid line
plt.xlabel("$X$")
plt.ylabel("$Y$")
xmin = -0.5
xmax = 10.5
ymin = -5
ymax = 105
plt.axis([xmin, xmax, ymin, ymax])
plt.legend(loc = "upper left")
plt.savefig("myplot.pdf")
```

# Generating Histograms #

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(size = 1000) # vect x
plt.hist(x)
```

```python
# Normalizing the histogram
plt.hist(x, density = True)
```

```python
# Providing Location for the Bins
# BINS - equal width areas (wide-ways) on a histogram. 
plt.hist(x, density = True, bins = np.linspace(-5, 5, 21)) # Twenty bins between -5 and +5. 21 points generated.

# More details on what is going on: 

#------*-------*-------# Stars are at start and end of bin. so for 2 points, you have only one bin.
# 21 points end up with only 20 bins. 

```

**GAMMA DISTRIBUTION** - is a continuous probability density funciton taht starts at 0 and goes all the way to positive infinity. 

```python
#100,000 samples to get a smooth histogram. 
x = np.random.gamma(2,3,100000)
x
plt.hist(x, bins = 30);
```

```python
plt.hist(x, bins = 30, density = True); # Normalized
```

```python
plt.hist(x, bins = 30, cumulative = True); # What percentage of the total number of cases accumulated each month
# and how much of the outbreak had taken place. 
```

```python
plt.hist(x, bins = 30, cumulative = True, density = True, histtype = "step");
```

**LOTS OF SUBPLOTS**

```python
plt.figure()
plt.subplot(221) # rows collumns and third argument gives the plot number (so you define 2 x 2 array every time 
                 # and see how it fills up. Left to right, top to bottom. 
plt.hist(x, bins = 30); # 1
plt.subplot(222)
plt.hist(x, bins = 30, density = True); # 2
plt.subplot(223)
plt.hist(x, bins = 30, cumulative = True); # 3
plt.subplot(224)
plt.hist(x, bins = 30, cumulative = True, density = True, histtype = "step"); # 4
```
