```python
from simncoins import simncoins as sc
from matplotlib import pyplot as plt

nsims = 50_000
ncoins = 100

heads_arr = sc(nsims, ncoins)
print(heads_arr)

plt.hist(heads_arr, bins= ncoins+1, range=(-0.5, ncoins+0.5))
plt.xlabel("Number of Heads")
plt.ylabel("Frequency of Occurence")
plt.title(str(ncoins) + " Coinflips")
plt.savefig(str(ncoins) + " coinflips.png")
```
